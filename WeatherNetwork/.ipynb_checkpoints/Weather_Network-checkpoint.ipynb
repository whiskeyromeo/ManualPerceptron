**1.) What are the inputs and outputs for this problem**

The inputs for this problem are the hours for each. The hour ($ X_A $) is used to establish the weights and the output of the linear activation function is compared to the desired output, which is the temperature of the given day. The output of the training is a set of weights( $ W_A,W_{n+1} $ )  which are then combined in a linear equation( $ X_B = W_AX_A + W_{n+1} $ ) and used in predicting the temperature( $ X_B $ ) at future points based on a given hour ( $ X_A $ ).

**2.) What should be the activation function of your decision unit? Why did you choose it?**

Based on the need for an output of continuous real values, I chose a linear activation function, which takes as input a net value established by applying two randomly generated weights ( $ W_A, W_{n+1} $ ) against the input hour ( $ X_A $ ) the function below:
$$ net = W_AX_A + X_{n+1} $$
and generates an output
$$ out = net $$ 
The error is then generated by subtracting the output from the desired output( $ d $ ), which is the actual temperature from that time on a given day.
$$ error = d - out $$
The error is then multiplied by the alpha rate to establish the rate of learning, which is used to update the set of weights. 

I found to get a better level of accuracy, this approach required the training of at least two perceptrons, which in turn output weights representative of their respective inputs ( hour ranges).

**5.) Report the number of iterations, the learning rate, and any data pre-processing steps you chose. Clearly explain why you selected these values.**

For this project I found no significant benefit in normalizing the data, so no significant preprocessing was done. The learning rate I chose was 0.01 as I ran into errors generated via Python's interpretation of the significantly small error values which were attained when trying larger alphas. I chose 1000 iterations as the learning rate as the error at that point seemed to be close to the point of inflection. Reduction of the error still occurs beyond that point, however I felt it was a good marker as it seemed to be near the point at which the rate of change in error dropped in the training on hours 9-13.

**6.) Based on the learning trend of the decision unit, would you be able to predict if the decision unit would improve its prediction if more iterations were allowed?**

It does not seem that the accuracy of the prediction would increase greatly, however the reduction in error would probably continue to some point. In order to get a reduction of any significant magnitude, I would most likely need to increase the amount of neurons which I am training on. The most accurate position may even be such that each hour is compared directly with its direct neighbors, in order to account for the local minima which arise at certain points. 

Any error reduction which could be achieved would be pertinent to the current training data, however due to the nature of the weather, it may be assumed that any predictions made based on a few days would not necessarily always carry precisely into the future. 

**7.) Could the error be further reduced using a neural network (opposed to a single decision unit)? If so, discuss how and why these methods would reduce the error.**

Based on the results demonstrated in the tables below, it seems that using more than a single decision unit does have a positive impact on reducing the overall error rate. Due to the variations across the hours during the day, slight valleys and peaks are not well accounted for when using a single line to fit the curve. In order to increase accuracy, we need to account for these variations in the data and try to establish a gradient which is more suited to the data at hand. It seems that for this dataset, one of the easiest methods is to increase the number of decision units. 

Adding layers to the network may also have a positive effect on error reduction, but the constraining point is still the number of decision units utilized in the network.

## Code

```python
from weather_project_code.data import prep_weather_data_for_graph, zipTrainingList
from weather_project_code.perceptron import Perceptron
from weather_project_code.graphlib import generate_comparison_dataframe, plot_weather_data, plot_error, display_table_for_iteration
import random
from IPython.display import display
%matplotlib inline
```

```python
masterTrainingList = prep_weather_data_for_graph(relative_path="./Project3_data/",file_list=["train_data_1.txt","train_data_2.txt","train_data_3.txt"])
training_data = zipTrainingList(masterTrainingList)

train1 = [x for x in training_data if x[0] <= 9]
train2 = [x for x in training_data if x[0] >= 9]

dimensions = 1
lin_weights = [random.random() for i in range(dimensions+1)]

plot_weather_data(masterTrainingList, title="Weather data days 1-3")
```

```python
iterations = 1000
alpha = 0.01
p = Perceptron(ite=iterations, alpha=alpha)
lin_res1 = p.train(train1, weights=lin_weights, activation="linear")
lin_res2 = p.train(train2, weights=lin_weights, activation="linear")
```

```python
plot_error(lin_res1[2::], title="Error vs Iterations, Hours 5-9")
plot_error(lin_res2[2::], title="Error vs Iterations, Hours 9-13")
```

```python
lin1_weights = lin_res1[-1]["weights"]
lin2_weights = lin_res2[-1]["weights"]
# Generate predictions for day 4 based off of the last training iteration
predictions = p.generatePredictions(lin1_weights, lin2_weights)

masterTrainingList.append(predictions)
plot_weather_data(masterTrainingList, title="Weather Data days 1-3 with day 4 Prediction")
```

```python
iterations = 1000
alpha = 0.01
p = Perceptron(ite=iterations, alpha=alpha)
lin_res3 = p.train(training_data, weights=lin_weights, activation="linear")
plot_error(lin_res3[2::], title="Error vs Iterations, Hours 5 - 13")

lin3_weights = lin_res3[-1]["weights"]
lin3_predictions = p.generatePredictions(lin3_weights, [])
```
