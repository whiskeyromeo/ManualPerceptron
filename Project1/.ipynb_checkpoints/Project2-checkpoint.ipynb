## Randomly select the sample from the normalized student list

- Since the data is randomly generated, this part is not necessary, but this approach allows for alternate datasets following a similar structure to be passed in via the `retrieve_students` method below.
- The data is randomly selected and then normalized using the `separate_normalized_students` method.


```python
from project_code.fileWorker import retrieve_students
import random
import numpy as np
import matplotlib.pyplot as plt
from project_code.data import normalize_student_data, separate_normalized_students, separate_and_normalize_students
from project_code.perceptron import Perceptron
from project_code.graphlib import plot_normalized_data, plot_animated_data

%matplotlib notebook
import random

# Retrieve the student data from a file
students = retrieve_students('./data.txt')
# Generate a normalized list of student data
student_list = normalize_student_data(students)
# Create a training set
student_train = random.sample(student_list, 2000)
student_test = random.sample(student_list, 4)
```

## Scenario A : Height Only

```python
student_height_list = [[x[0],x[2]] for x in student_list]
student_train = random.sample(student_height_list, 2000)
weights = [random.random() for i in range(2)]
```

```python
iterations=100
p = Perceptron(ite=iterations, alpha=0.2)
weights = [random.random() for i in range(2)]

hard_results = p.train(student_train, weights=weights, activation="hard")

male_students, female_students = separate_normalized_students(student_train,gender_idx=1)

modified_weights = hard_results[-1][1]
print(modified_weights)

y_intercept = -(modified_weights[0]/modified_weights[1])
fig4 = plt.figure("Scenario A: Height Only, Hard Activation, Alpha=0.1")
anim = plot_animated_data(fig4, male_students, female_students, hard_results,1)
plt.show()
```

```python
iterations=100
p = Perceptron(ite=iterations, alpha=0.1)
weights = [random.random() for i in range(2)]

soft_results = p.train(student_train, weights=weights, activation="soft")
male_students, female_students = separate_normalized_students(student_train,gender_idx=1)

weights_list = [x[1] for x in soft_results]


modified_weights = hard_results[-1][1]

y_intercept = -(modified_weights[0]/modified_weights[1])
fig5 = plt.figure("Scenario A: Soft Activation, Alpha=0.1")
anim = plot_animated_data(fig5, male_students, female_students, soft_results,1)
plt.show()
```

## Scenario B: Height and Weight

```python
# Generate a normalized list of student data
student_list = normalize_student_data(students)
# Create a training set
student_train = random.sample(student_list, 2000)
student_test = random.sample(student_list, 4)

weights = [random.random() for i in range(3)]
male_students, female_students = separate_normalized_students(student_train)
print(weights)
```

```python
iterations=10
p = Perceptron(ite=iterations, alpha=0.01)
hard_results = p.train(student_train, weights=weights, activation="hard")
fig = plt.figure()
anim = plot_animated_data(fig, male_students, female_students, hard_results)
plt.show()
```

```python
iterations=10
p = Perceptron(ite=iterations, alpha=0.5)
hard_results = p.train(student_train, weights=weights, activation="hard")
fig2 = plt.figure("Height x Weight : High alpha")
anim = plot_animated_data(fig2, male_students, female_students, hard_results)
plt.show()
```

```python
iterations=500
p = Perceptron(ite=iterations, alpha=0.1)
hard_results = p.train(student_train, weights=[0.,0.,0.], activation="soft")
fig3 = plt.figure("Scenario B : Height x Weight, alpha= 0.1, Soft")
anim = plot_animated_data(fig3, male_students, female_students, hard_results)
plt.show()
```
