## Analysis

### Scenario A  - Hard Activation

#### Overview


The Perceptron model trained using the hard activation utilizes the following logic for determining how to modify the weights. 

$$ out = \begin{cases} 1 & if\ {\sum_{i=1}^{n} w_ix_i  + w_{n+1}} \ge 0\\ 0 & if\ {\sum_{i=1}^{n} w_ix_i  + w_{n+1}} < 0 \end{cases}
$$

The resulting output is then subtracted from the desired output in order to determine an error value of either 0 or 1 which then multiplied by the learning rate in order to determine the effect of the error on the change in the weights. This change is the amount learned for each iteration through the dataset used for training in each iteration. The *hard* aspect of the function comes from the fact that each modification to the weights is of a set value, $ 1 * alpha $ or $ 0 * alpha $. Here the learning rate plays a key role, as the amount of learning done in each cycle is determined by the adjustment the learning rate does to the error. A large learning rate will cause a larger adjustment in the weights and may ultimately lead to a greater degree of correction required by the system in future steps.

#### Splits

In all cases tested, with a high variance in the alpha value tested, none of the split methods consistently produced a higher accuracy than any of the other methods. This was true for both 1-dimensional and 2-dimensional data. This may be attributed to the model never converging on a delineating line due to the amount of overlap in the data, resulting in the gradient constantly shifting until termination of the final iteration as an error of $ epsilon < 10*e^-5 $ is never reached. The set of weights achieved at the end of the training with the hard activation function is not necessarily the best set of weights which was encountered throughout the training, it is simply the last modification to the set. This is an important difference from the soft activation method, in which the last set of weights is closer to what might be optimal considering the overlap. 

### Scenario B - Soft Activation


#### Overview

The method of soft activation we chose to implement was that of the sigmoid function, which is defined as follows:

$$
out = \frac{1}{1+\exp(-\mathbf{w}^T\mathbf{x}_i)}
$$

The resulting output follows the same pattern as mentioned above for the hard activation function, however the range of error produced by the method enables a smaller movement towards the reduction of error, as the values produced by the function are within a range of 0 to 1.  If a point is misclassified, as with the hard activation method, the difference between the expected output and the output produced by the function when multiplied by the learning rate will determine how much needs to be learned by the weights in the current step. The lower value produced by the soft activation function enables the perceptron to move towards a minimal point of error in smaller steps than is possible using the hard activation function.


#### Splits

The results obtained using the soft activation function are very similar to those described above for Scenario A. Ultimately the same conclusions are reached regarding the behavior of the Perceptron, though there are some differences which are worth noting regarding the overall accuracy of the method and the general behavior of the model trained with the soft activation function.

There was no discernable benefit to a single split over another, as was the case with the hard activation approach. Again this may be attributed to the significant overlap in the data points between male and female students. The main difference between the methods is apparent in the accuracy calculation of the model, which in the case of the soft approach are close to +/- 8% more accurate in the best case than the best results found using the hard approach.
As in scenario A, the epsilon termination point is never reached and the training continues until the end of the iterations. The weights produced are a result of the small modifications to the error by the sigmoid function which ultimately causes convergence to a minimal error rate even in the presence of overlap.


### Summary

Due to the large overlap in male and female student data, there was not a significant difference to be found in the behavior of either the hard or soft perceptron models using the various train/test splits. This may be attributed to the highly overlapping nature of the data which prevented an ideal split from being obtained and ultimate convergence on a particular gradient thus unachievable. The noticeable difference between the behavior of the methods is how they converge on their ultimate error values, with a much larger variation in the weights produced by the perceptron trained using hard activation than one trained using soft activation. 

As mentioned above, in hard activation the selection of the final weights with overlapping data is essentially one of chance. The final weight set produced is that which was selected when the iterations terminated. In the case of soft activation, there is a more gradual movement towards the smallest error possible while still having misclassified points. Since we only have a limited number of weights from which to operate, the ultimate set is one in which the error is the least possible given the selected learning rate. The learning rate and the number of iterations chosen will determine how closely the soft activation function may converge on gradient with minimal possible error. 


It is important to note the structure of the data used in these analyses and address potential points of error which might be introduced. The method **runSplits** takes a number of parameters, one of the most important of which is the boolean **shuffle**. This parameter takes the dataset from the data.txt file and shuffles the data while maintaining an even distribution of male and female students in the train/test set, which is then split via the index point for the relative ratios. We found this approach to be very important as it enabled us to use training and test sets with different data members to compare the behavior of the relative systems. There is another method commented out in the last cell, **runAllSplits** which has the same base functionality as runSplits but runs hard and soft activation functions on the same dataset. This further corroborated the findings above and helped reduce potentially significant points of error which might have erroneously led us to the conclusion that one type of split performed consistently better than another. 


By using the accompanying animation notebook, you can explore how the training is performed on various sizes of data. With few enough points, both hard and soft activation will find the gradient which can be used to classify the students. Generally, with a low enough epsilon, soft activation will converge to the more central point between the students however, whereas hard activation will lead to the stopping point whenever the first separating gradient is achieved.
